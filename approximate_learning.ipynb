```python
from smartcab.run import run_sims
from smartcab.approx_learn_agent import ApproxLearnAgent

df_approx = run_sims(100, 100, ApproxLearnAgent)
df_approx.to_csv('approx_agent_results.csv')
df_approx.describe()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

results = {'original': pd.DataFrame.from_csv("learning_agent_results.csv"),
           'optimistic': pd.DataFrame.from_csv("optimistic_agent_results.csv"),
           'perfect': pd.DataFrame.from_csv("perfect_agent_results.csv"),
          'approximate': df_approx}

final_panel = pd.Panel.from_dict(results)
for column in df_approx.columns:
    sns.boxplot(data=final_panel.minor_xs(column))
    plt.title(column)
    plt.show()
    plt.close()
```
